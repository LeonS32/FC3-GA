```python
Adjacency = nx.adjacency_matrix(G).todense()
Adjacency =np.where(Adjacency == 0, np.inf, Adjacency)
print(Adjacency)
```

```python
class GeneticAlgorithms:
    def __init__(self, Npop, Nit, max_length, Adjacency, start_node, end_node):
        self.Npop = Npop  # Number of chromosomes
        self.Nit = Nit  # Number of iterations
        self.max_length = max_length  # Maximum chromosome length
        self.Adjacency = np.array(Adjacency)  # Graph adjacency matrix
        self.start_node = start_node  # Fixed start node
        self.end_node = end_node  # Fixed end node
        self.population = [self.random_chromosome() for _ in range(Npop)]

    def random_chromosome(self):
        if random.random() < 0.1:  # 10% chance to create an elementary solution
            return [self.start_node, self.end_node]
        length = random.randint(2, self.max_length - 2)  # Subtract 2 for the fixed start and end nodes
        middle_nodes = [random.randint(0, len(self.Adjacency) - 1) for _ in range(length)]
        return [self.start_node] + middle_nodes + [self.end_node]

    def crossover(self, parent1, parent2):
        # Single-point crossover, excluding start and end nodes
        min_len = min(len(parent1), len(parent2)) - 2
        if min_len < 1:
            return parent1[:], parent2[:]  # Avoid crossover if too few intermediate nodes

        crossover_point = random.randint(1, min_len)
        child1 = [self.start_node] + parent1[1:crossover_point+1] + parent2[crossover_point+1:-1] + [self.end_node]
        child2 = [self.start_node] + parent2[1:crossover_point+1] + parent1[crossover_point+1:-1] + [self.end_node]
        return child1, child2

    def mutate(self, chromosome):
        # Mutation can only change intermediate nodes
        if len(chromosome) > 2:  # Ensure there's at least one intermediate node to mutate
            mutation_type = random.choice(['insert', 'delete', 'change'])
            position = random.randint(1, len(chromosome) - 2)  # Avoid mutating the first and last nodes
            if mutation_type == 'insert' and len(chromosome) < self.max_length:
                node = random.randint(0, len(self.Adjacency) - 1)
                chromosome.insert(position, node)
            elif mutation_type == 'delete' and len(chromosome) > 3: #must be more than 3 nodes to not stay only with the star y end nodes
                chromosome.pop(position)
            elif mutation_type == 'change':
                node = random.randint(0, len(self.Adjacency) - 1)
                chromosome[position] = node

    def fitness(self, chromosome):
        # Calculate the total path cost of a chromosome
        total_cost = 0
        for i in range(len(chromosome) - 1):
            cost = self.Adjacency[chromosome[i], chromosome[i + 1]]
            if cost == np.inf:
                return np.inf  # Penalize infeasible paths heavily
            total_cost += cost
        return total_cost

    def select(self, population, fitness_scores):
        fitness_scores = np.array(fitness_scores)

        # Invert fitness scores to make lower cost more likely. Avoid division by zero or infinity.
        probabilities = 1 / (fitness_scores + 1e-6)
        probabilities[np.isinf(fitness_scores)] = 0  # Set probabilities to 0 where fitness is infinity

        # Normalize probabilities to sum to 1
        total_probability = np.sum(probabilities)
        if total_probability > 0:
            probabilities /= total_probability
        else:
            # If total_probability is 0 (e.g., if all valid fitness_scores are infinite), distribute uniformly
            probabilities = np.ones_like(fitness_scores) / len(fitness_scores)

        # Select indices based on computed probabilities
        selected_indices = np.random.choice(len(population), size=self.Npop, replace=True, p=probabilities)
        return [population[i] for i in selected_indices]

    def evolve(self): #creates a new population
        for _ in range(self.Nit):
            fitness_scores = [self.fitness(chrom) for chrom in self.population]
            self.population = self.select(self.population, fitness_scores)
            new_population = []
            while len(new_population) < self.Npop:
                parent1, parent2 = random.sample(self.population, 2)
                child1, child2 = self.crossover(parent1, parent2)
                self.mutate(child1)
                self.mutate(child2)
                new_population.extend([child1, child2])

            self.population = new_population[:self.Npop]

    def run(self): #Runs all the functions and select the best chromosome
        self.evolve()
        final_fitness = [self.fitness(chrom) for chrom in self.population]
        best_index = np.argmin(final_fitness)
        best_chromosome = self.population[best_index]
        best_fitness = final_fitness[best_index]

        if best_fitness == np.inf:
            self.run()
        else:
            print("Best chromosome:", best_chromosome)
            print("Best fitness (path cost):", best_fitness)
```

```python
ga = GeneticAlgorithms(Npop=1000, Nit=500, max_length=len(Adjacency), Adjacency=Adjacency, start_node=0, end_node=9)
ga.run()
```
